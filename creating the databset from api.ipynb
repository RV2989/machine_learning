```python
import pandas as pd
```

```python
import requests

url = "https://omgvamp-hearthstone-v1.p.rapidapi.com/info"

headers = {
	"X-RapidAPI-Key": "1b13f13b11msh70478792b574b0bp184e41jsn9e038ef47dd1",
	"X-RapidAPI-Host": "omgvamp-hearthstone-v1.p.rapidapi.com"
}

response = requests.get(url, headers=headers)
data=response.json()
print(response.json())
```

```python
classes_data=data['classes']
sets_data=data['sets']
standard_data=data['standard']
max_length = max(len(classes_data), len(standard_data),len(sets_data))
classes_data = classes_data + [''] * (max_length - len(classes_data))
standard_data = standard_data + [''] * (max_length - len(standard_data))
sets_data = sets_data + [''] * (max_length - len(sets_data))
df=pd.DataFrame({'classes':classes_data,'standard':standard_data,'sets':sets_data})
```
